```python
#2 Задание

df["delivery_cost_per_unit"] = df.apply(lambda row: row["highway_cost"] / sum(product["quantity"] for product in row["products"]), axis=1)
tariff = df.groupby("warehouse_name")["delivery_cost_per_unit"].mean().reset_index()


list = []


for index, row in df.iterrows():
    for product in row["products"]:
        product_name = product["product"]
        quantity = product["quantity"]
        price = product["price"]
        income = quantity * price
        expenses = tariff.loc[tariff["warehouse_name"] == row["warehouse_name"], "delivery_cost_per_unit"].values[0] * quantity
        profit = income + expenses
        

        list.append({
            "product": product_name,
            "quantity": quantity,
            "income": income,
            "expenses": expenses,
            "profit": profit
        })


sum_df = pd.DataFrame(list)


sum_df.groupby("product").sum().reset_index()

```

```python
#3 Задание
df["delivery_cost_per_unit"] = df.apply(lambda row: row["highway_cost"] / sum(product["quantity"] for product in row["products"]), axis=1)
tariff = df.groupby("warehouse_name")["delivery_cost_per_unit"].mean().reset_index()

order_list = []

for index, row in df.iterrows():
    order_id = row["order_id"]
    total_profit = sum(product["price"] * product["quantity"] for product in row["products"]) + row["highway_cost"]
    warehouse_name = row['warehouse_name']
    order_list.append({
        "order_id": order_id,
        "order_profit": total_profit,
        "warehouse_name": warehouse_name
    })

order_df = pd.DataFrame(order_list)

#print(order_df)

avg_profit = order_df["order_profit"].mean()
#print("Средняя прибыль заказов:", avg_profit)
warehouse_profit = order_df.groupby('warehouse_name')['order_profit'].sum()
warehouse_profit
```

```python
#4 Задание
df["delivery_cost_per_unit"] = df.apply(lambda row: row["highway_cost"] / sum(product["quantity"] for product in row["products"]), axis=1)
tariff = df.groupby("warehouse_name")["delivery_cost_per_unit"].mean().reset_index()


sum_list = []


for _, row in df.iterrows():
    warehouse_name = row["warehouse_name"]
    for product in row["products"]:
        product_name = product["product"]
        quantity = product["quantity"]
        price = product["price"]
        expenses = tariff.loc[tariff["warehouse_name"] == warehouse_name, "delivery_cost_per_unit"].values[0] * quantity
        income = quantity * price + expenses

        
        sum_list.append({
            "warehouse_name": warehouse_name,
            "product": product_name,
            "quantity": quantity,
            'income':income
        })

sum_df = pd.DataFrame(sum_list)


sum_df = sum_df.groupby(['product','warehouse_name']).sum().reset_index()

sum_df = sum_df.join(warehouse_profit,on = 'warehouse_name')

sum_df['percent_profit_product_of_warehouse'] = sum_df['income'] / sum_df['order_profit'] * 100

sum_df.sort_values(by = 'warehouse_name', ascending = False)
```

```python
#5 Задание 
import pandas as pd


sorted_df = sum_df.sort_values(by='percent_profit_product_of_warehouse', ascending=False)


sorted_df['accumulated_percent_profit_product_of_warehouse'] = sorted_df.groupby('warehouse_name')['percent_profit_product_of_warehouse'].transform(pd.Series.cumsum)

sorted_df = sorted_df.sort_values(by='warehouse_name', ascending=False)
sorted_df
```

```python
#6 Задание

def category(accumulated_percent):
    if accumulated_percent <= 70:
        return 'A'
    elif 70 < accumulated_percent <= 90:
        return 'B'
    else:
        return 'C'


sorted_df['category'] = sorted_df['accumulated_percent_profit_product_of_warehouse'].apply(category)
sorted_df



```
